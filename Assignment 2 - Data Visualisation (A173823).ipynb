## Histogram

```python
weightDf = df['Weight (kg)']
weightDf
```

```python
plt.figure()
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.hist(weightDf, bins=5, edgecolor='black',facecolor='blue', alpha=0.5)
plt.show()
```

## Boxplot

```python
plt.ylabel('Weight (kg)')
plt.boxplot(weightDf, showmeans=True, vert=False)
```

## Analysis Section

Mean, Minimum and Maximum values based on weight

```python
print('Mean: ', weightDf.mean())
```

```python
print('Minimum value: ', weightDf.min())
```

```python
print('Maximum value: ', weightDf.max())
```

The histogram and boxplot graph above is about weight of laptop in kg. The maximum of weight is 1.65 kg while the minimum of weight is 1.1 kg. The median of weight is approximately 1.44 kg based on the boxplot graph above. Which means, the lower quartile (Q1) in interquatile range of weight between 1.1 kg to 1.44 kg. While the upper quartile (Q3) in interquatile range of weight between 1.44 kg to 1.65 kg.<br>
As shown in the graph, there are 2 outliers which above the maximum and bottom of minimum. That is because some of the data values are too great than the others which can be shown on boxplot graph.

## Scatterplot

```python
weightDf = df['Weight (kg)']
# processorDf = df['Processor']
warrantyDf = df['Warranty (Month)']
```

```python
plt.figure()
plt.scatter(weightDf,warrantyDf)
plt.xlabel('Weight (kg)')
plt.ylabel('Warranty (Month)')
plt.show()
```

The scatterplot graph above shows a weak, positive, non-linear association between weight of laptop in kg and warranty in month. There appear to be some outliers in the graph. The most obvious outlier is 1.1 kg of weight which distinct from other data.<br>
